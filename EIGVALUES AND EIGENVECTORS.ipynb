```python
import numpy as np
```

```python
B = ([[3,1],[0,2]])
print("matrix B:")
print(B)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(B)
```

```python
print("\nEigenvalues:")
print(eigenvalues)
```

```python
print("\nEigenvectors:")
print(eigenvectors)
```

```python
for i in range (len(eigenvalues)):
    Bv = np.dot(B,eigenvectors[:,i])
    Iv = eigenvalues[i] * eigenvectors[:,i]
    print(f"\nverification for eigenvalues{eigenvalues[i]}:")
    print("B * V", Bv)
    print("B * 3",Iv)
```
